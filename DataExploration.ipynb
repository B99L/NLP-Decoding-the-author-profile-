```python
import os
import xml.etree.ElementTree as ET
from collections import defaultdict
import re
import concurrent.futures

# Specify the top-level directory
dir_path = "C:/Users/nevio/OneDrive - ZHAW/Dokumente/NLP/Project 4"

# Specify the number of workers to use
num_workers = 6  # Change to the number of workers you want to use

# Function to clean the author's name
def clean_author_name(name):
    # Convert the name to lowercase
    name = name.lower()
    # Remove '(Text)' (in any case) and anything after it
    name = re.sub(r'\(text\).*', '', name, flags=re.I)
    # Remove anything after a comma
    name = name.split(',')[0]
    # Remove 'von' in any case
    name = re.sub(r'\b[vV][oO][nN]\b', '', name)
    return name.strip()  # Remove leading/trailing white space

# Function to process a single XML file
def process_file(file_path):
    # Initialize a dictionary to hold author text counts for this file
    author_text_counts = defaultdict(int)
    # Parse the XML file
    tree = ET.parse(file_path)
    # Get the root of the XML document
    root = tree.getroot()
    # Find all 'au' tags in the document
    authors = root.findall('.//au')
    if authors:
        for au in authors:
            # Ignore empty tags and 'interview' tags
            if au.text and au.text.strip() != '' and not au.text.lower().startswith('interview:'):
                # Clean the author's name
                author_name = clean_author_name(au.text)
                # Increment the text count for this author
                author_text_counts[author_name] += 1
    return author_text_counts

# Initialize a dictionary to hold author text counts
total_author_text_counts = defaultdict(int)

# Iterate over each directory in the top-level directory
for dirpath, dirnames, filenames in os.walk(dir_path):
    # Get a list of all XML files in the current directory
    xml_files = [os.path.join(dirpath, name) for name in filenames if name.endswith('.xml')]
    # Use a ProcessPoolExecutor to process the XML files concurrently
    with concurrent.futures.ProcessPoolExecutor(max_workers=num_workers) as executor:
        for author_text_counts in executor.map(process_file, xml_files):
            # Add the author text counts from this file to the total counts
            for author, count in author_text_counts.items():
                total_author_text_counts[author] += count

# Print the text counts for each author
for author, count in total_author_text_counts.items():
    print(f'{author}: {count}')
```

```python
import os
import xml.etree.ElementTree as ET
from collections import defaultdict
import re
import concurrent.futures
import multiprocessing

# Specify the top-level directory
dir_path = "C:/Users/nevio/OneDrive - ZHAW/Dokumente/NLP/Project 4"

# Specify the number of workers to use
num_workers = multiprocessing.cpu_count()

# Function to clean the author's name
def clean_author_name(name):
    # Convert the name to lowercase
    name = name.lower()
    # Remove '(Text)' (in any case) and anything after it
    name = re.sub(r'\(text\).*', '', name, flags=re.I)
    # Remove anything after a comma
    name = name.split(',')[0]
    # Remove 'von' in any case
    name = re.sub(r'\b[vV][oO][nN]\b', '', name)
    return name.strip()  # Remove leading/trailing white space

# Function to process a single XML file
def process_file(file_path):
    # Initialize a dictionary to hold author text counts for this file
    author_text_counts = defaultdict(int)
    try:
        # Parse the XML file
        tree = ET.parse(file_path)
        # Get the root of the XML document
        root = tree.getroot()
        # Find all 'au' tags in the document
        authors = root.findall('.//au')
        if authors:
            for au in authors:
                # Ignore empty tags and 'interview' tags
                if au.text and au.text.strip() != '' and not au.text.lower().startswith('interview:'):
                    # Clean the author's name
                    author_name = clean_author_name(au.text)
                    # Increment the text count for this author
                    author_text_counts[author_name] += 1
    except ET.ParseError as e:
        print(f'ParseError reading file {file_path}: {e}')
    except Exception as e:
        print(f'Unexpected error processing file {file_path}: {e}')
    return author_text_counts


# Initialize a dictionary to hold author text counts
total_author_text_counts = defaultdict(int)

# Iterate over each directory in the top-level directory
for dirpath, dirnames, filenames in os.walk(dir_path):
    # Get a list of all XML files in the current directory
    xml_files = [os.path.join(dirpath, name) for name in filenames if name.endswith('.xml')]

    # Use a ProcessPoolExecutor to process the XML files concurrently
    # Use a ThreadPoolExecutor to process the XML files concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=num_workers) as executor:
        future_to_file = {executor.submit(process_file, file): file for file in xml_files}
        for future in concurrent.futures.as_completed(future_to_file):
            file = future_to_file[future]
            try:
                author_text_counts = future.result()
            except Exception as exc:
                print('%r generated an exception: %s' % (file, exc))
            else:
                # Add the author text counts from this file to the total counts
                for author, count in author_text_counts.items():
                    total_author_text_counts[author] += count


# Print the text counts for each author
for author, count in total_author_text_counts.items():
    print(f'{author}: {count}')
```

```python
import os
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor
from fuzzywuzzy import fuzz
from tqdm import tqdm
import pandas as pd

top_authors = ["oliver meiler", "stefan häne", "markus brotschi"]

def get_text_from_txt(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        return f.read().strip()

def find_top_author_text(xml_file):
    # Open and parse the XML file
    with open(xml_file, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f.read(), 'lxml')

    # Search for the 'au' tag
    au_tag = soup.find('au')
    if au_tag is None:
        return None

    # Check if any of the top authors is in the 'au' tag
    for top_author in top_authors:
        if fuzz.token_set_ratio(top_author, au_tag.text.lower()) > 80:
            # The corresponding txt file
            txt_file = xml_file.replace('.xml', '.txt')

            # Return the author, file path and text
            return {
                'Author': top_author,
                'Filepath': txt_file,
                'Text': get_text_from_txt(txt_file)
            }

    return None

# Directory where the XML files are located
base_dir = "C:/Users/nevio/OneDrive - ZHAW/Dokumente/NLP/Project 4"

# List to hold the paths of all XML files
xml_files = []

# Traverse through all subdirectories
for root, dirs, files in os.walk(base_dir):
    for file in files:
        # If the file is an XML file, add its path to xml_files
        if file.endswith(".xml"):
            xml_files.append(os.path.join(root, file))

# List to hold the dataframes for each file
dfs = []

# Use a ThreadPoolExecutor to process the XML files concurrently
with ThreadPoolExecutor(max_workers=4) as executor:
    # Initialize the progress bar
    with tqdm(total=len(xml_files), desc="Processing files") as pbar:
        # Process each XML file
        for result in executor.map(find_top_author_text, xml_files):
            # Update the progress bar
            pbar.update()

            # If the result is not None, add it to dfs
            if result is not None:
                dfs.append(pd.DataFrame([result]))

# Concatenate all the dataframes
df = pd.concat(dfs, ignore_index=True)

# Save the dataframe to a CSV file
df.to_csv('author_texts.csv', index=False)
```

```python
from bs4 import BeautifulSoup
import os
from tqdm import tqdm
import pandas as pd

papers = ["NZZ", "BAZ", "BU", "BZ", "TA"]
base_dir = "C:/Users/nevio/OneDrive - ZHAW/Dokumente/NLP/Project 4"
dftags = []

# Get total number of files
total_files = sum([len(files) for r, d, files in os.walk(base_dir)])

# Set up tqdm progress bar
pbar = tqdm(total=total_files)

for paper in papers:
    for year in range(1996, 2024):
        year_dir = os.path.join(base_dir, paper, str(year))

        # Go through each sub-directory in the year directory
        for subdir, dirs, files in os.walk(year_dir):
            for file in files:
                # Only process XML files
                if file.endswith(".xml"):
                    file_path = os.path.join(subdir, file)
                    with open(file_path, 'r', encoding='utf-8') as f:
                        contents = f.read()
                        soup = BeautifulSoup(contents, 'lxml')
                        au_tags = soup.find_all('au')

                        # Add each author to the DataFrame
                        for au in au_tags:
                            author = au.string if au else ""
                            dftags.append({'Filepath': file_path, 'Author': author})

                    # Update progress bar
                    pbar.update(1)

# Create a pandas DataFrame from the data
dftags = pd.DataFrame(dftags)
pbar.close()

# Print all unique author entries
print(dftags['Author'].unique())
```
